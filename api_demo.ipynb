```python
import pandas as pd
import json
import requests
```

```python
url = 'https://pokeapi.co/api/v2/pokemon/gyarados'
response = requests.get(url)
```

```python
response.text
```

```python
response_json = json.loads(response.text)
response_json
```

```python
df = pd.json_normalize(response_json, record_path = 'stats')
df['pokemon'] = 'gyarados'
df.head()
```
